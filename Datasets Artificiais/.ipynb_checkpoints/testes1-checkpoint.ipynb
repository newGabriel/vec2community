```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering as spectral
import hdbscan
```

```python
def purity_score(y,y_pre):
    m = {}
    for i in range(len(y)): # traverse the results matrix and store the objects that belong to each cluster on a map
        if y_pre[i] in m: 
            m[y_pre[i]].append(i)
        else:
            m[y_pre[i]] = [i]
     
    tot = 0
    
    for i in m: # check the most common class in each of the clusters
        c = {}
        mx = 0 # number of members of the most common class in the cluster
        for j in m[i]: 
            if y[j] in c:
                c[y[j]] += 1
            else:
                c[y[j]] = 1
            mx = max(mx,c[y[j]])
        tot += mx
    return tot/len(y)
```

```python
df = pd.read_csv("banana.txt",header=None)
df.head()
```

```python
X = np.array(df[[0,1]])
X
```

```python
mx = 0
for i in range(1,101):
    km = KMeans(n_clusters=i)
    km.fit(X)
    p = purity_score(df[2],km.labels_)
    c = purity_score(km.labels_,df[2])
    mx = max(mx,(2*p*c)/(p+c))
mx
```

```python
res = pd.DataFrame()
```

```python
res["banana"]=[mx]
res
```
